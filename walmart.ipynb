```python
def holiday_name(date) :
  if date in [pd.Timestamp('2010-02-12'),pd.Timestamp('2011-02-11'),pd.Timestamp('2012-02-10') ] :
    return 'Super Bowl'
  elif date  in [pd.Timestamp('2010-09-10'),pd.Timestamp('2011-09-09'),pd.Timestamp('2012-09-07') ] :
    return 'Labor Day'
  elif date in [pd.Timestamp('2010-11-26'),pd.Timestamp('2011-11-25'),pd.Timestamp('2012-11-23') ] :
    return 'Thanksgiving'
  elif date in [pd.Timestamp('2010-12-31'),pd.Timestamp('2011-12-30'),pd.Timestamp('2012-12-28') ] :
    return 'Christmas'
  else :
    return 'No Holiday'

df['Holiday_name'] = df['Date'].apply(holiday_name)
```

Analysis of Holiday Impact and Seasonal Trends
The in-depth analysis of weekly sales by year and month, focusing on holiday seasons, yielded the following strong insights:
Overall Yearly Performance: 2011 recorded the best overall sales performance compared to other years in the dataset.
Major Holiday Impact: The impact of major holidays on weekly sales was significantly high compared to non-holiday weeks, though the relative importance varied:
Thanksgiving: Showed a consistently high and strong impact on sales across all years.
Super Bowl: Its impact was notably stronger in 2012 compared to other years.
Seasonal Trends (Monthly): When sales are aggregated monthly, the influence of major holidays becomes clear:
November (Thanksgiving): Ranks as the second-highest sales month.
December (Christmas): Is the highest-selling month overall, driven by the Christmas holiday season.
Strategic Conclusion: Holiday seasons are a critical driver of sales peaks. While non-holiday weeks can sometimes show high sales, the dedicated holiday weeks, particularly Thanksgiving and Christmas, consistently deliver the highest sales and must be accurately modeled.

```python
from sklearn.model_selection import train_test_split

target = 'Weekly_Sales'

features = {
    'lag_1', 'lag_2', 'lag_3', 'lag_4',
    'rolling_mean', 'rolling_std',
    'rolling_mean_4', 'rolling_std_4',
    'days_from_last_holiday', 'days_from_next_holiday',
    'year', 'month', 'WeekOfYear', 'DayOfWeek'
}

df_model = df.dropna(subset = list(features) + [target])
```

```python
x = df_model[list(features)]
y = df_model[target]

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , shuffle = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd # Ensure pandas is imported if not globally available

# Identify categorical columns that are currently strings
categorical_cols = ['month', 'DayOfWeek']

# Perform one-hot encoding on training and test sets
x_train_encoded = pd.get_dummies(x_train, columns=categorical_cols, drop_first=True)
x_test_encoded = pd.get_dummies(x_test, columns=categorical_cols, drop_first=True)

# Align columns - crucial for consistent feature sets between train and test
train_cols = x_train_encoded.columns
test_cols = x_test_encoded.columns

missing_in_test = set(train_cols) - set(test_cols)
for c in missing_in_test:
    x_test_encoded[c] = 0

missing_in_train = set(test_cols) - set(train_cols)
for c in missing_in_train:
    x_train_encoded[c] = 0

x_test_encoded = x_test_encoded[train_cols] # Ensure order and presence of columns

model = RandomForestRegressor(n_estimators =100 , random_state =42)
model.fit(x_train_encoded , y_train)

y_pred = model.predict(x_test_encoded)

mse = mean_squared_error(y_test , y_pred)
print("MSE IS : " , mse)
```

```python
importance = pd.Series(model.feature_importances_ , index = x_train_encoded.columns)

imporatnces = importance.sort_values(ascending = False)

plt.figure(figsize = (10,8))
importance.plot(kind='bar')
plt.title("Feature Importance ")
plt.ylabel("Importance")
plt.xlabel("Features")
plt.show()
```

```python
def rolling_slope(series_segment, window=4):
    slopes = [np.nan] * len(series_segment)
    for i in range(window, len(series_segment)):
        y_values = series_segment.iloc[i-window : i].values
        if np.isnan(y_values).any():
            slopes[i] = np.nan
            continue
        x_values = np.arange(window)
        coef = np.polyfit(x_values, y_values, 1)[0]
        slopes[i] = coef
    return pd.Series(slopes, index=series_segment.index)

df['rolling_slope'] = df.groupby('Store')['lag_1'].transform(rolling_slope, window=4)

df['week_before_holiday'] = df['lag_1'] / df['lag_2'].replace(0, np.nan) - 1
df['week_before_holiday'] = df['week_before_holiday'].replace([np.inf, -np.inf], np.nan)

df['year_over_year_change'] = df['lag_1'] / df['lag_52'].replace(0, np.nan) - 1
df['year_over_year_change'] = df['year_over_year_change'].replace([np.inf, -np.inf], np.nan)

df['is_holiday'] = df['Holiday_Flag'].fillna(0).astype(int)

df['rolling_slope'] = df['rolling_slope'].fillna(0)
df['week_before_holiday'] = df['week_before_holiday'].fillna(0)
df['year_over_year_change'] = df['year_over_year_change'].fillna(0)
```
